## Data Visualization

### Histogram of Age of Passengers

```r
library(ggplot2)
library(readxl)
Titanic_Data <- read_excel("~/Fall Semester 2018/Statistical Analysis with R/Assignments/Assignment 12/Titanic Data.xlsx")
Titanic_Data$Survived = factor(Titanic_Data$Survived)
Titanic_Data$Pclass = factor(Titanic_Data$Pclass)
```

```r
ggplot(data=Titanic_Data) + geom_histogram(mapping=aes(x=Age, fill=Sex))
```

```r
ggplot(data=Titanic_Data) + geom_histogram(mapping=aes(x=Age, fill=Survived))
```

```r
ggplot(data=Titanic_Data) + geom_histogram(mapping=aes(x=Age, fill=Pclass))
```

### Gender Distribution of Passengers

```r
ggplot(data=Titanic_Data) + geom_bar(mapping=aes(x=Sex, fill=Survived))
```

### Count Graph of Sex v. Survived

```r
ggplot(data=Titanic_Data) + geom_count(mapping=aes(x=Survived, y=Sex))
```

### Passenger Class Distribution

```r
ggplot(data=Titanic_Data) + geom_bar(mapping=aes(x=Pclass, fill=Sex))
```

```r
ggplot(data=Titanic_Data) + geom_bar(mapping=aes(x=Pclass, fill=Survived))
```
